## CountVectorizer Scikit-learn
###### Ekstraksi fitur dengan cara menghitung jumlah frekuensi kemunculan tiap kata pada teks


https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html?highlight=countvectorizer#sklearn.feature_extraction.text.CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# CONTOH dengan sampel kalimat
sampleText = ['saya suka belajar', 'saya suka bekerja', 'saya belajar', 'saya suka suka suka']
```

```python
sample_vectorizer = CountVectorizer()
sample_resultVectorizer = sample_vectorizer.fit_transform( sampleText )

# Cetak hasil dari ekstraksi fitur menggunakan CountVectorizer
pandas.DataFrame(sample_resultVectorizer.toarray(), index=sampleText, columns=sample_vectorizer.get_feature_names())
```

## K-Nearest Neighbors (KNN) Scikit-learn
###### Klasifikasi teks menggunakan KNN


https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html#sklearn.neighbors.KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# CONTOH dengan sampel kalimat hasil CountVectorizer
printableDataLatih = sample_resultVectorizer.toarray().tolist()

# Memberikan sampel label untuk kalimat secara berurutan
sampleLabel = ['POSITIF', 'NEGATIF', 'POSITIF', 'NEGATIF']
for (index, label) in enumerate( sampleLabel ):
    printableDataLatih[index].append( label )


# Preview isi dataLatih
pandas.DataFrame(printableDataLatih, index=sampleText, columns=sample_vectorizer.get_feature_names()+['label'])
```

```python
# Set KNN untuk mencari 3(tiga) ketetanggaan terdekatnya
knn = KNeighborsClassifier(n_neighbors=3)

# Fit model latih (parameter : clean_text, label)
knn.fit( sample_resultVectorizer.toarray().tolist(), sampleLabel )
```

```python
# Prediksikan model (knn) dengan vektor [0, 1, 1, 1] == 'saya suka belajar'
dataTes = sample_resultVectorizer.toarray().tolist()[0]

dataTes
```

```python
# Lihat probabilitas prediksinya
pandas.DataFrame(knn.predict_proba([ dataTes ]), index=['PROBABILITAS'], columns=['NEGATIF', 'POSITIF'])
```

```python
# Prediksi label dari variabel dataTes berdasarkan model latih
print(knn.predict([ dataTes ]))
```
